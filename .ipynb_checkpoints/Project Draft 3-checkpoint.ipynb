#### Part 2 Fill NaN values in the training set

```python
#Drop the Id column and SalePrice column in the training set. 
#The Id column will not be used in the model. And SalePrice will be the y for the training.
train=train.drop(['SalePrice'],axis=1)
train=train.drop(['Id'],axis=1)
train.head()
```

```python
#Drop the Id column in the test set
#The Id column will not be used in the model.
test=test.drop(['Id'],axis=1)
test.head()
```

```python
#This code shows all numerical columns in the training set.
train._get_numeric_data().columns
train.head()
```

```python
#For convenience I use this code to check how many null values are in these numerical (literally) columns
train[train._get_numeric_data().columns].isnull().sum()
```

```python
#It seems that only two columns listed above have NaN values. 
#Fill NaN of the LotFrontage Column
train['LotFrontage']=train['LotFrontage'].fillna(train['LotFrontage'].mean())
#the 7th row in the LotFrontage column was NaN. I want to check whether it is filled
train
```

```python
#Based on previous results, GarageYrBlt column has 81 NaN. 
#This column looks like a numerical one but it is not. So fill NaN with 'None'
#NaN simply indictes: there is no garage at this property.
#Later this column will be treated using get_dummmies function as other categorical columns
train['GarageYrBlt']=train['GarageYrBlt'].fillna('None')
```

```python
#Now all NaN values in columns with numbers are filled. Now deal with NaN in other columns
#In these columns, NaN means 'None', as this property does not have something
#One function will get the job done. 
train=train.fillna('None')
train.head()
```
