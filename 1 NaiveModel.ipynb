### Model 1: Naive Model

#### Part 1 Basics

```python
#Basic Imports
import numpy as np
import pandas as pd
```

```python
#Load the data
train=pd.read_csv('./input/train.csv')
train
```

```python
#This code will be used later in part 5
y=train['SalePrice']
```

```python
#Load the data
test=pd.read_csv('./input/test.csv')
test
```

```python
testid=test['Id']
```

```python
#Drop the Id column and SalePrice column in the training set. 
#The Id column will not be used in the model. And SalePrice will be the y for the training.
train=train.drop(['SalePrice'],axis=1)
train=train.drop(['Id'],axis=1)
train.head()
```

```python
#Drop the Id column in the test set
#The Id column will not be used in the model.
test=test.drop(['Id'],axis=1)
test.head()
```

```python
#Before handle NaN and dummy variables, concatenate the train and test set. 
#So that deal with NaN and dummy variables will be done together. 
alldata=pd.concat([train,test],axis=0)
alldata.head()
```

#### Part 2 Fill NaN values 

```python
#This code shows all numerical columns in the training set.
alldata._get_numeric_data().columns
```

```python
#For convenience I use this code to check how many null values are in these numerical (literally) columns
alldata[alldata._get_numeric_data().columns].isnull().sum()
```

```python
#Fill NaN using the mean of that column (by default ignored NaN)
alldata['LotFrontage']=alldata['LotFrontage'].fillna(alldata['LotFrontage'].mean())
alldata['MasVnrArea']=alldata['MasVnrArea'].fillna(alldata['MasVnrArea'].mean())
```

```python
#For these columns, NaN simply means '0' as this property does not have a basement at all
alldata['BsmtFinSF1']=alldata['BsmtFinSF1'].fillna(0)
alldata['BsmtFinSF2']=alldata['BsmtFinSF2'].fillna(0)
alldata['BsmtUnfSF']=alldata['BsmtUnfSF'].fillna(0)
alldata['TotalBsmtSF']=alldata['TotalBsmtSF'].fillna(0)
```

```python
#For these columns, NaN means 0 since there is no bathroom at the basement (or no basement at all)
alldata['BsmtFullBath']=alldata['BsmtFullBath'].fillna(0)
alldata['GarageArea']=alldata['GarageArea'].fillna(0)
```

```python
#Based on previous results, GarageYrBlt column has 159 NaN. 
#NaN simply indictes: there is no garage at this property.
#These NaN values will be filled using the 0 to reflect the fact that there is no garage
alldata['GarageYrBlt']=alldata['GarageYrBlt'].fillna(0)
```

```python
#For these columns, NaN means 0 since there is no garage
alldata['GarageCars']=alldata['GarageCars'].fillna(0)
alldata['BsmtHalfBath']=alldata['BsmtHalfBath'].fillna(0)
```

```python
#Now all NaN values in columns with numbers are filled. Now deal with NaN in other columns
#In these columns, NaN means 'None', as this property does not have something
#One function will get the job done. 
alldata=alldata.fillna('None')
alldata.head()
```

```python
#Check if there is still an NaN value
alldata.isnull().sum()
```

#### Part 3 Create dummy variables

```python
alldata=pd.get_dummies(alldata)
alldata.head()
```

#### Part 4 Train/Test Split (Back to the state before the concat)

```python
train=alldata[:train.shape[0]]
train
```

```python
test=alldata[train.shape[0]:]
test
```

#### Part 5 The Naive Model

```python
X_train=train
y_train=y #See part 1
```

```python
X_train
```

```python
y_train
```

```python
#This null simply predict that the price would be the mean of y_train for all properties
y_test_naivemodel=pd.Series([np.mean(y_train)] * (y_train.size-1))
y_test_naivemodel.head()
```

```python
np.count_nonzero(y_test_naivemodel)
```

#### Part 6 Evaluate the Performance

```python
#The performance will be evaluated by submitting a file to Kaggle 
#Kaggle uses RMSE for performance evaluation
#The remaining codes in this part: create a Kaggle submission file
sub1=pd.DataFrame(testid)
sub1.head()
```

```python
sub2=pd.DataFrame(y_test_naivemodel,columns=['SalePrice'])
sub2.head()
```

```python
sub=pd.concat([sub1,sub2],axis=1)
sub.head()
```

```python
sub.to_csv('sub_NaiveModel.csv',index=False)
```
